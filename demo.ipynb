```python
from collections import OrderedDict
from bokeh.charts import output_file
from bokeh.palettes import Spectral11
from bokeh.plotting import figure, show

from bokeh.sampledata.iris import flowers
from bokeh.charts import Scatter

import numpy as np
import pandas as pd
import tables
import urllib.request
```

```python
f = urllib.request.urlopen('http://research.rods.pitt.edu/data.csv')
file = f.read()
chartdata = pd.read_csv(file)
chartdata
```

```python
chartdata = chartdata[chartdata['Infection_state'] == 'NEWLY_SICK']
chartdata = chartdata.groupby(['Day'])['Count'].aggregate(sum)
chartdata = chartdata.to_frame("Count")
chartdata = chartdata.reset_index()
chartdata

```

```python
p = Scatter(chartdata, x='Day', y='Count',
            title="Newly Sick Per Day", legend="top_left",
            xlabel="Day", ylabel="Newly Sick")

output_file("nspd.html")

show(p)
```
