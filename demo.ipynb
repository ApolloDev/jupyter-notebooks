```python
from collections import OrderedDict
from bokeh.charts import output_file
from bokeh.palettes import Spectral11
from bokeh.plotting import figure, show

from bokeh.sampledata.iris import flowers
from bokeh.charts import Scatter

import numpy as np
import pandas as pd
import tables
```

```python
line_listing = pd.read_csv('http://research.rods.pitt.edu/line_listing.csv')
line_listing
```

```python
newly_sick = line_listing[line_listing['infection_state'] == 'NEWLY_SICK']

newly_sick_per_day = newly_sick.groupby(['simulator_time', 'integer_age'])['count'].aggregate(sum)
newly_sick_per_day = newly_sick_per_day.to_frame("count")
newly_sick_per_day = newly_sick_per_day.reset_index()
newly_sick_per_day
```

```python
plot = Scatter(newly_sick_per_day, x='simulator_time', y='count',
            title="Newly Sick Per Day", legend="top_left",
            xlabel="Day", ylabel="Newly Sick")
output_file("nspd.html")
show(plot)
```

```python
newly_sick = line_listing[line_listing['infection_state'] == 'NEWLY_SICK']

newly_sick_per_day_by_sex = newly_sick.groupby(['simulator_time', 'sex'])['count'].aggregate(sum)
newly_sick_per_day_by_sex = newly_sick_per_day_by_sex.to_frame("count")
newly_sick_per_day_by_sex = newly_sick_per_day_by_sex.reset_index()
newly_sick_per_day_by_sex
```

```python
plot = Scatter(newly_sick_per_day_by_sex, x='simulator_time', y='count', color='sex',
            title="Newly Sick per Day by Sex", legend="top_left",
            xlabel="Day", ylabel="Newly Sick")
output_file("nspdbs.html")
show(plot)
```

```python
newly_sick = line_listing[line_listing['infection_state'] == 'NEWLY_SICK']

newly_sick_per_day_by_sex_and_state = newly_sick.groupby(['simulator_time', 'sex','location_admin1'])['count'].aggregate(sum)
newly_sick_per_day_by_sex_and_state = newly_sick_per_day_by_sex_and_state.to_frame("count")
newly_sick_per_day_by_sex_and_state = newly_sick_per_day_by_sex_and_state.reset_index()

newly_sick_females_per_day_by_state = (newly_sick_per_day_by_sex_and_state
                                       [newly_sick_per_day_by_sex_and_state.sex.str.contains("F") == True])
newly_sick_females_per_day_by_state
```

```python
plot = Scatter(newly_sick_females_per_day_by_state, x='simulator_time', y='count', color='location_admin2',
            title="Newly Sick Females per Day by State", legend="top_left",
            xlabel="Day", ylabel="Newly Sick")
output_file("nspdbs.html")
show(plot)
```

```python
newly_sick = line_listing[line_listing['infection_state'] == 'NEWLY_SICK']

newly_sick_per_day_by_state_and_county \
    = newly_sick.groupby(['simulator_time', 'sex','location_admin1',"location_admin2"])['count'].aggregate(sum)
    
newly_sick_per_day_by_state_and_county = newly_sick_per_day_by_state_and_county.to_frame("count")
newly_sick_per_day_by_state_and_county = newly_sick_per_day_by_state_and_county.reset_index()

newly_sick_females_per_day_by_state_and_county = (newly_sick_per_day_by_state_and_county
    [newly_sick_per_day_by_state_and_county.sex.str.contains("F") == True])
    
newly_sick_females_per_day_in_state_1 = (newly_sick_females_per_day_by_state_and_county
    [newly_sick_females_per_day_by_state_and_county.location_admin1.str.contains("State1") == True])

newly_sick_females_per_day_in_state_1
```

```python
p = Scatter(newly_sick_females_per_day_in_state_1, x='simulator_time', y='count', color='location_admin2',
            title="Newly Sick Females per Day by Counties in State 1", legend="top_left",
            xlabel="Day", ylabel="Newly Sick")
output_file("nspdbs.html")
show(p)
```

```python
newly_sick = line_listing[line_listing['infection_state'] == 'NEWLY_SICK']

newly_sick_per_day_by_state_and_county \
    = newly_sick.groupby(['simulator_time', 'sex','location_admin1',"location_admin2", "integer_age"])['count'].aggregate(sum)
    
newly_sick_per_day_by_state_and_county = newly_sick_per_day_by_state_and_county.to_frame("count")
newly_sick_per_day_by_state_and_county = newly_sick_per_day_by_state_and_county.reset_index()

newly_sick_females_per_day_by_state_and_county = (newly_sick_per_day_by_state_and_county
    [newly_sick_per_day_by_state_and_county.sex.str.contains("F") == True])
    
newly_sick_females_per_day_in_state_1 = (newly_sick_females_per_day_by_state_and_county
    [newly_sick_females_per_day_by_state_and_county.location_admin1.str.contains("State1") == True])

newly_sick_females_per_day_in_state_1_county_1 = (newly_sick_females_per_day_by_state_and_county
    [newly_sick_females_per_day_by_state_and_county.location_admin2.str.contains("County1") == True])


newly_sick_females_per_day_in_state_1_county_1

bins = [0, 5, 18, 25, 65, 100]

def func(row):
    return pd.cut([row['integer_age']], bins=bins, labels=["infant","school_age","college_age","working_age","twilight_age"])[0]

newly_sick_females_per_day_in_state_1_county_1['age_group'] = newly_sick_females_per_day_in_state_1_county_1.apply(func, axis=1)
newly_sick_females_per_day_in_state_1_county_1= newly_sick_females_per_day_in_state_1_county_1.groupby(['simulator_time', 'age_group'])['count'].aggregate(sum)
newly_sick_females_per_day_in_state_1_county_1 = newly_sick_females_per_day_in_state_1_county_1.to_frame("count")
newly_sick_females_per_day_in_state_1_county_1 = newly_sick_females_per_day_in_state_1_county_1.reset_index()
newly_sick_females_per_day_in_state_1_county_1
```

```python
p = Scatter(newly_sick_females_per_day_in_state_1_county_1, x='simulator_time', y='count', color='age_group',
            title="fix me", legend="top_left",
            xlabel="Day", ylabel="Newly Sick")
output_file("nspdbs.html")
show(p)
```
